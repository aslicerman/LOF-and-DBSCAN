# Anomaly Detection Example with DBSCAN in Python

DBSCAN  yoğunluk tabanlı bir kümeleme algoritmasıdır. Bu algoritmanın temel ilkesi, yoğun bir alanda esas örnekleri bulması ve kümeler oluşturmak için örnekleri bu esas örnekler çevresinde gruplandırmasıdır. Buna göre düşük yoğunluklu bir alandaki numuneler aykırı değerler haline gelir. 
 
Scikit-learn API, bu algoritma için DBSCAN sınıfını sağlar. 

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from numpy import random, where
import matplotlib.pyplot as plt
```

**Rastgele bir örneklem kümesi oluşturuyoruz.**

```python
random.seed(13)
x, _ = make_blobs(n_samples=200, centers=1, cluster_std=.3, center_box=(20, 5))
```

**Veri setini grafik ile görselleştirerek kontrol ediyoruz.**

```python
plt.scatter(x[:,0], x[:,1])
plt.show()
```

**Modeli ve anomali testini tanımlama**

Modeli Scikit-learn API'nin DBSCAN sınıfını kullanarak tanımlayacağız. Sınıfın argümanlarında 'eps' ve 'min_sample' tanımlayacağız. 'Eps' argümanı iki örnek arasındaki mesafedir ve 'min_samples' örnek sayısıdır.

```python
dbscan = DBSCAN(eps = 0.28, min_samples = 20)
print(dbscan) 
```

**Modeli x veri kümesine fit ediyoruz ve tahmin verilerini fit_predict () yöntemiyle alıyoruz.**

```python
pred = dbscan.fit_predict(x)
```

**Sonra, negatif çıktıları aykırı değerler olarak çıkarıyoruz.**

```python
anom_index = where(pred == -1)
values = x[anom_index]
```

**Son olarak, anormallikleri bir renkle vurgulayarak sonuçları bir grafikte görselleştiriyoruz.**

```python
plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0], values[:,1], color='r')
plt.show()
```

# Anomaly Detection Example with Local Outlier Factor in Python

The Local Outlier Factor, gözlem verilerindeki anormallikleri tespit etmek için bir algoritmadır. Her örneğin yerel yoğunluk puanının ölçülmesi ve puanlarının ağırlıklandırılması algoritmanın ana konseptidir. Algoritma, numunenin puanını komşularıyla karşılaştırarak, düşük yoğunluklu öğeleri verilerdeki anormallikler olarak tanımlar.

Temel olarak bir nesnenin yerel yoğunluğu komşularının yerel yoğunlukları ile karşılaştırılarak, benzer yoğunluktaki bölgeler ve komşularından önemli ölçüde daha düşük yoğunluğa sahip noktalar belirlenmektedir.  Bunlar aykırı değerler olarak kabul edilir.

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs
from numpy import quantile, where, random
import matplotlib.pyplot as plt
```

**Rastgele bir örneklem kümesi oluşturuyoruz.**

```python
random.seed(1)
x, _ = make_blobs(n_samples=200, centers=1, cluster_std=.3, center_box=(10,10))
```

**Veri setini grafik ile görselleştirerek kontrol ediyoruz.**

```python
plt.scatter(x[:,0], x[:,1])
plt.show()
```

**Modeli ve anomali testini tanımlama**

Modeli Scikit-learn API'nin LocalOutlierFactor sınıfını kullanarak tanımlayacağız. Argümanlarda tahminci sayısını ve kirlilik değerini belirleyeceğiz. Contamination, bir veri kümesindeki aykırı değerlerin oranını tanımlar.

```python
lof = LocalOutlierFactor(n_neighbors=20, contamination=.03)
```

**Modeli x veri kümesine fit ediyoruz ve tahmin verilerini fit_predict () yöntemiyle alıyoruz.**

```python
y_pred = lof.fit_predict(x)
```

**Negatif çıktıları aykırı değer olarak çıkarıyoruz.**

```python
lofs_index = where(y_pred==-1)
values = x[lofs_index]
```

**Son olarak, anormallikleri bir renkle vurgulayarak sonuçları bir grafikte görselleştiriyoruz.**

```python
plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0],values[:,1], color='r')
plt.show()
```

### Anomaly detection with scores

**İkinci yöntemde, contamination argümanını ayarlamadan modeli tanımlıyoruz.**

```python
model = LocalOutlierFactor(n_neighbors=20)
```

**Modeli x veri kümesine fit ediyoruz, ardından örnek puanını çıkarıyoruz.**

```python
model.fit_predict(x)
lof = model.negative_outlier_factor_ 
```

**Daha sonra, quantile fonksiyonunu kullanarak puanlardan eşik değerini elde ediyoruz. Burada, puan değerlerinin en düşük yüzde 3'ünü anormallikler olarak alıyoruz..**

```python
thresh = quantile(lof, .03)
print(thresh)
-1.8191482960907037
```

**Eşik değerini karşılaştırarak anormallikleri çıkarıyoruz ve öğelerin değerlerini belirliyoruz.**

```python
index = where(lof<=thresh)
values = x[index]
```

**Son olarak, anormallikleri bir renkle vurgulayarak sonuçları görselleştiriyoruz.**

```python
plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0], values[:,1], color='r')
plt.show()
```

**Yukarıdaki her iki yöntemde de aynı sonucu aldık. Bunlardan herhangi birini analizinizde kullanabilirsiniz. Eşik veya kirlilik değeri, daha aşırı durumları filtrelemek için değiştirilebilir.**
